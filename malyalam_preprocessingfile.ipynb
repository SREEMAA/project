```python
import re
import pandas as pd
#emoji removal
def remove_emoji(string):
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return emoji_pattern.sub(r'', string)


df = pd.read_csv("/content/drive/MyDrive/mtech project final/Nonh1.csv",encoding='UTF-16')
df
          
          
def preprocess(comment):
            text = remove_emoji(comment)
            #removing english words
            text=re.sub("[a-zA-Z]+", "",text)
            #removing special characters
            text=re.sub("[<>/:.!,&$#@~^%*()_'}{?]+", "", text)
            #removing numbers
            text=re.sub("[0-9]", "", text) 
          
            #removing white spaces
            text=re.sub("  ", "", text)
            #removing white spaces at leading and ending
            text=re.sub("^\s+|\s+?$", "", text)
            #stop words removal
            stopwords =["ഏത്","എന്തിന്റെ","ഇത്ര","ഏറെ","ഇതേ","എന്ന","വേറെ","മതി","എല്ലാ","നിങ്ങളെ","വളരെ","എവിടെ","എപ്പോൾ","ഇന്ന്","ഉള്ള","ഇന്ന്","മറ്റു","ഈ","ഇത്","പോലെ","തന്റെ","വരെ","മാത്രം","എന്നാൽ","മുമ്പ്","തന്റെ","ഇനി","എങ്കിൽ","അല്ലെങ്കിൽ","കുറിച്ചു","പിന്നെ","എന്നത്","അന്ന്","എന്ത്"]
            querywords = text.split()
            resultwords  = [word for word in querywords if word.lower() not in stopwords]
            text = ' '.join(resultwords)  
            print(text)
            df.iat[i,0] =text

  

for i in range(len(df)):
    value = df.iat[i,0]
    preprocess(value)
    
df
# saving the dataframe
df.to_csv('/content/drive/MyDrive/mtech project final/fil2.csv')

df = df.dropna()
```
